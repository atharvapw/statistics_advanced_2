Question 1: What is hypothesis testing in statistics?

Answer:
Hypothesis testing is a statistical method used to make decisions or inferences about a population based on sample data. It involves formulating two competing hypotheses — the null hypothesis (H₀) and the alternative hypothesis (H₁) — and using sample data to determine whether there is enough evidence to reject the null hypothesis. The goal is to assess whether observed data significantly differ from what we would expect under the null hypothesis.


Question 2: What is the null hypothesis, and how does it differ from the alternative hypothesis?

Answer:

Null Hypothesis (H₀): It assumes that there is no effect or no difference in the population, and any observed variation is due to random chance.

Alternative Hypothesis (H₁ or Ha): It states that there is an effect or a difference in the population.
Example:
H₀: The average height of students = 170 cm
H₁: The average height of students ≠ 170 cm

The null hypothesis is tested for possible rejection, while the alternative hypothesis represents the conclusion we draw if we reject H₀.

Question 3: Explain the significance level in hypothesis testing and its role in deciding the outcome of a test.

Answer:
The significance level (α) is the probability of rejecting the null hypothesis when it is actually true (Type I error). Common values are 0.05 or 0.01.
It acts as a threshold for decision-making:

If the p-value ≤ α, reject H₀ (evidence supports H₁).

If the p-value > α, fail to reject H₀ (insufficient evidence).
It determines how strict the test is — smaller α values make rejection of H₀ harder.

| Error Type  | Meaning                    | Example                          |
| ----------- | -------------------------- | -------------------------------- |
| Type I (α)  | Rejecting true H₀          | Declaring innocent person guilty |
| Type II (β) | Failing to reject false H₀ | Declaring guilty person innocent |

Question 5: What is the difference between a Z-test and a T-test? Explain when to use each.

Answer:

| Test       | Used When                      | Population Std. Dev. Known? | Sample Size    |
| ---------- | ------------------------------ | --------------------------- | -------------- |
| **Z-Test** | Population variance is known   | Yes                         | Large (n > 30) |
| **T-Test** | Population variance is unknown | No                          | Small (n ≤ 30) |


The Z-test assumes a normal distribution with known σ, while the T-test uses the sample standard deviation and follows a t-distribution.

```python
# Question 7: Implement hypothesis testing using Z-statistics

import numpy as np
from scipy import stats

# Given sample data
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

# Hypothesis: μ = 50
mu = 50
sample_mean = np.mean(sample_data)
sample_std = np.std(sample_data, ddof=1)
n = len(sample_data)

# Z-statistic
z = (sample_mean - mu) / (sample_std / np.sqrt(n))

# Two-tailed test
p_value = 2 * (1 - stats.norm.cdf(abs(z)))

print(f"Sample Mean: {sample_mean:.2f}")
print(f"Z-Statistic: {z:.3f}")
print(f"P-Value: {p_value:.4f}")

if p_value < 0.05:
    print("Reject the null hypothesis.")
else:
    print("Fail to reject the null hypothesis.")
```

```python
# Question 8: Simulate normal data and calculate 95% confidence interval

import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Simulate data
np.random.seed(0)
data = np.random.normal(loc=100, scale=15, size=100)

# Confidence interval (95%)
mean = np.mean(data)
sem = stats.sem(data)
conf_interval = stats.t.interval(0.95, len(data)-1, loc=mean, scale=sem)

# Plot
plt.hist(data, bins=15, edgecolor='black', alpha=0.7)
plt.title("Normal Distribution Data")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.axvline(mean, color='red', linestyle='dashed', linewidth=2, label=f"Mean = {mean:.2f}")
plt.legend()
plt.show()

print(f"Mean = {mean:.2f}")
print(f"95% Confidence Interval = {conf_interval}")
```

```python
# Question 9: Calculate Z-scores and visualize

import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Sample dataset
data = np.random.normal(50, 5, 100)

# Calculate Z-scores
z_scores = stats.zscore(data)

# Plot histogram
plt.hist(z_scores, bins=15, edgecolor='black', alpha=0.7)
plt.title("Z-Scores of Standardized Data")
plt.xlabel("Z-Score")
plt.ylabel("Frequency")
plt.show()

print("Z-Scores represent how many standard deviations each data point is from the mean.")
```
